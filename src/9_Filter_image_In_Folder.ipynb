```python
path_tuan_hung = "/media/damvantai/data/Documents/datasets/face/vietnam/tuan_hung/"
```

```python
path_tuan_hung_test = '/media/damvantai/data/Documents/datasets/face/vietnam/tuan_hung_test/'
```

```python
path_yes = "/media/damvantai/data/Documents/datasets/face/vietnam/nhandien_tuanhung/"
path_no = "/media/damvantai/data/Documents/datasets/face/vietnam/nhandien_khongco_tuanhung/"
```

```python
face_encodings = [] # face encodings BUI ANH TUAN
path = ""
```

```python
for filename in os.listdir(path_tuan_hung):
    image = face_recognition.load_image_file(path_tuan_hung + filename)
    face_encoding = face_recognition.face_encodings(image)[0]
    face_encodings.append(face_encoding)
```

```python
min_distance_all = 111
```

```python
for file in os.listdir(path_tuan_hung_test):
    min_distance_all = 111
    image = face_recognition.load_image_file(path_tuan_hung_test + file)
    face_encoding = face_recognition.face_encodings(image)
    if len(face_encoding) == 0:
        path = path_tuan_hung_test + file
        shutil.copy2(path, path_no)
    else:
        for i in range(len(face_encoding)):
            distance = face_recognition.face_distance(face_encodings, face_encoding[i])
            min_distance = np.min(distance)
            if min_distance_all > min_distance:
                min_distance_all = min_distance
                
        if min_distance_all < 0.6:
            path = path_tuan_hung_test + file
            shutil.copy2(path, path_yes)
        else:
            path = path_tuan_hung_test + file
            shutil.copy2(path, path_no)
```

```python
img = cv2.imread(path)
```
